```python
import itertools
import random
import sys

letter_bag = ['A','A','A','A','A','A','A','A','A','B',
              'B','C','C','D','D','D','D','E','E','E',
              'E','E','E','E','E','E','E','E','E','F',
              'F','G','G','G','H','H','I','I','I','I',
              'I','I','I','I','I','J','K','L','L','L',
              'L','M','M','N','N','N','N','N','N','O',
              'O','O','O','O','O','O','O','P','P','Q',
              'R','R','R','R','R','R','S','S','S','S',
              'T','T','T','T','T','T','U','U','U','U',
              'V','V','W','W','X','Y','Y','Z']


def getTiles(myTiles): #mengambil tiles/huruf
    myTiles += random.sample(letter_bag, 7)
    return myTiles


def printTiles(myTiles): #menampilkan tiles dan score masing-masing tiles
    tiles = ""
    score = ""
    for letter in myTiles:
        tiles += letter + "  "
        thisScore = getScore(letter)
        if thisScore > 9:
            score += str(thisScore) + " "
        else:
            score += str(thisScore) + "  "

    print("\nHuruf : " + tiles)
    print("Nilai : " + score)


scoresFile = open('drive/MyDrive/AI_Projek/scores.txt') #membuka file score
Scores = [] #membaca file score dan memasukkan ke scores
for line in scoresFile:
    line = line.split()
    letter = line[0]
    score = int(line[1])
    Scores.append([letter,score])
scoresFile.close()


def getScore(letter): #mengambil score dari sebuah huruf
    for item in Scores:
        if item[0] == letter:
            return item[1]


myTiles = [] #inisiasi tiles
getTiles(myTiles)
printTiles(myTiles)

def score_word(wordzz):
    counter_1 = 0
    for letter in Scores:
        for letter_1 in wordzz:
            if letter[0] == letter_1:#checks if the user input character = the letter stored in the "Scores"
                counter_1 += letter[1]#add the score of the alphabet to "counter_1" occuring in user input
                                                   
    return counter_1   #returning the score of the word


letterCombos = [] #inisiasi kommbinasi kata
def getLetterCombos(letter): #fungsi mencari semua kemungkinan kombinasi kata
        letters = myTiles
        
        for length in range(2, len(letters)+1):
            listTuples = list( itertools.permutations(list(letters), length) )
            letterCombos.append( set(listTuples) )
        return letterCombos 


getLetterCombos(myTiles) #memanggil fungsi

wordz=[]
for letter in letterCombos: #menggabungkan letter menjadi kata
  for char in letter:
    strz = ""
    str1 = strz.join(char)
    wordz.append(str1)


dictFile = open("drive/MyDrive/AI_Projek/diction.txt") #membaca kamus

Dict = []
for line in dictFile:
  line = line.split()
  wordZ = line[0]
  Dict.append([wordZ])
dictFile.close()


#for word in list(Dict): #mereduksi kamus agar hanya mengandung kata yang terdidi dari 2-7 huruf
  #for line in word:
    #if len(line) > 7:
      #Dict.remove(word)
#with open("drive/MyDrive/AI_Projek/dictionary.txt", 'w') as writefile:
    #for line in Dict:
        #writefile.write(line)


words_3=[] #pencarian kata yang ada dikamus
for letter in wordz:
  for words in list(Dict):
    for line in words:
      if letter == line:
        words_3.append([letter])
print("Kata-kata yang dapat anda bentuk :  ")
print(words_3)


word_score =0 #mencari kata tertinggi
list1 = [0]
high_score = " "
for wordddd in words_3:
  for wordx in wordddd:
    word_score = score_word(wordx)
    list1.append(word_score)
    if list1[0]<=word_score:
      list1[0] = word_score
      high_score = wordx
if list1[0] ==0:
  print("Tidak ada kata yang dapat dibentuk")
else:
  print("Kata " + high_score + " memiliki nilai tertinggi. Nilainya adalah " + str(list1[0]))    




```
